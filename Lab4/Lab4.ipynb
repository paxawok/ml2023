```python
#імпорт всього необхідного
import tensorflow as tf
from sklearn.datasets import make_classification
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error, roc_curve, auc
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# a. Створення тензорів

# a. Створення тензорів

# Створення скалярного тензору
scalar_tensor = tf.constant(2 * 3.14)
print("Скалярний тензор:")
print(scalar_tensor)

# Створення вектора
vector_tensor = tf.constant([i**2 for i in range(1, 4)])
print("\nВекторний тензор:")
print(vector_tensor)

# Створення матриці
angles = tf.constant([[0, 3.14 / 4], [3.14 / 2, 3 * 3.14 / 4]])
sin_cos_matrix = tf.math.sin(angles) + tf.math.cos(angles)
print("\nМатричний тензор:")
print(sin_cos_matrix)

# b. Операції з тензорами

# Операції додавання, віднімання, множення та ділення
tensor_a = tf.constant([[1, 2, 3], [3, 4, 5]])
tensor_b = tf.constant([[5, 6, 7], [7, 8, 9]])

addition_result = tensor_a + tensor_b
subtraction_result = tensor_a - tensor_b
multiplication_result = tensor_a * tensor_b
division_result = tf.divide(tensor_a, tensor_b)

print("\nРезультати операцій:")
print("Додавання:")
print(addition_result)
print("Віднімання:")
print(subtraction_result)
print("Множення:")
print(multiplication_result)
print("Ділення:")
print(division_result)

# Застосування складних математичних функцій
logarithmic_result = tf.math.log(tf.cast(tensor_a, dtype=tf.float32))
sqrt_result = tf.math.sqrt(tf.cast(tensor_a, dtype=tf.float32))

print("\nЗастосування складних математичних функцій:")
print("Логарифм:")
print(logarithmic_result)
print("Квадратний корінь:")
print(sqrt_result)

# c. Індексація та зрізи

# Звернення до конкретних елементів тензора
element_at_index = tensor_a[1, 0]
print("\nЗвернення до конкретного елемента:")
print(element_at_index)

# Зріз тензора з використанням різних виразів
tensor_slice = tensor_a[:, 1] + 10 * tensor_a[:, 0]
print("\nЗріз тензора:")
print(tensor_slice)

# d. Зміна форми тензорів

# Змінення розміру та форми тензора
reshaped_tensor = tf.reshape(tensor_a, (1, -1))
print("\nЗмінена форма тензора:")
print(reshaped_tensor)
```

```python
# a. Регресія

# Згенеруємо синтетичний набір даних
np.random.seed(1420)
X_regression = np.random.rand(1000, 1)  # Ознака
y_regression = 3 * X_regression + 4 + 0.1 * np.random.randn(1000, 1)  # Відповідна змінна

# Розбиття на тренувальний та тестовий набори
X_train_reg, X_test_reg, y_train_reg, y_test_reg = train_test_split(X_regression, y_regression, test_size=0.2)

# Створення моделі регресії
model_regression = tf.keras.Sequential([
    tf.keras.layers.Dense(units=1, input_shape=(1,), activation='linear')
])

# Компіляція моделі для регресії
model_regression.compile(optimizer='sgd', loss='mean_squared_error', metrics=['mse'])

# Навчання моделі
model_regression.fit(X_train_reg, y_train_reg, epochs=100, batch_size=16, validation_data=(X_test_reg, y_test_reg))

# Оцінка результатів для регресії
mse_regression = model_regression.evaluate(X_test_reg, y_test_reg)[1]
print("Mean Squared Error (MSE) for Regression:", mse_regression)

# Візуалізація результатів для регресії
y_pred_reg = model_regression.predict(X_test_reg)
plt.scatter(X_test_reg, y_test_reg, color='green', alpha=0.5, label='Actual')
plt.scatter(X_test_reg, y_pred_reg, color='black', alpha=0.5, label='Predicted')
plt.title('Regression Results')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()

#гістограма
plt.hist(y_test_reg - y_pred_reg, bins=30, color='blue', alpha=0.7, rwidth=0.85)
plt.title('Residuals Histogram for Regression')
plt.xlabel('Residuals')
plt.ylabel('Frequency')
plt.show()
```

```python
# b. Класифікація

# Згенеруємо синтетичний набір даних для класифікації
X_classification, y_classification = make_classification(n_samples=1000, n_features=20, n_classes=2, random_state=42)

# Розбиття на тренувальний та тестовий набори
X_train_cls, X_test_cls, y_train_cls, y_test_cls = train_test_split(X_classification, y_classification, test_size=0.2, random_state=42)

# Створення моделі класифікації
model_classification = tf.keras.Sequential([
    tf.keras.layers.Dense(units=32, input_shape=(20,), activation='relu'),
    tf.keras.layers.Dense(units=1, activation='sigmoid')
])

# Компіляція моделі для класифікації
model_classification.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Навчання моделі
model_classification.fit(X_train_cls, y_train_cls, epochs=50, batch_size=32, validation_data=(X_test_cls, y_test_cls))

# Оцінка результатів для класифікації
y_pred_cls = (model_classification.predict(X_test_cls) > 0.5).astype(int)
accuracy_cls = accuracy_score(y_test_cls, y_pred_cls)
conf_matrix_cls = confusion_matrix(y_test_cls, y_pred_cls)

print("Accuracy for Classification:", accuracy_cls)
print("Confusion Matrix for Classification:")
print(conf_matrix_cls)

# Візуалізація матриці плутанини
sns.heatmap(conf_matrix_cls, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix for Classification')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()

# Візуалізація ROC-кривої
fpr, tpr, thresholds = roc_curve(y_test_cls, model_classification.predict(X_test_cls))
roc_auc = auc(fpr, tpr)

plt.figure()
plt.plot(fpr, tpr, color='g', lw=2, label='ROC curve (area = {:.2f})'.format(roc_auc))
plt.plot([0, 1], [0, 1], color='black', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.show()

# Відобразіть дані в двохвимірному просторі (якщо ознак більше двох, використовуйте методи зменшення розмірності)
plt.scatter(X_test_cls[:, 0], X_test_cls[:, 1], c=y_test_cls, cmap='viridis', edgecolors='k', marker='o', s=50, label='Actual')
plt.title('Scatter Plot for Classification')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()

# Гістограма
plt.hist([y_test_cls.flatten(), y_pred_cls.flatten()], bins=30, color=['black', 'green'], alpha=0.7, rwidth=0.85, label=['Actual', 'Predicted'])
plt.title('Histogram for Classification')
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
